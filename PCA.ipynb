```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

```

```python
scaled_df = df1.copy()
scaler = StandardScaler()
scaled_features = scaler.fit_transform(scaled_df[['Total_Charges', 'Age_Ordinal', 'Length_of_Stay']])
scaled_df[['Total_Charges', 'Age_Ordinal', 'Length_of_Stay']] = scaled_features
scaled_df.head()
```

```python
pca = PCA(n_components=6)
principalComponents = pca.fit_transform(scaled_df)
```

```python
# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
plt.title("Variance Drop-off After the 2nd Component")
# plt.savefig("PCA_components.pdf")
```

```python
# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.1, color='black')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title("Scatter Plot of the PCA Model")
# plt.savefig("PCA_clusters.pdf")
```

```python
len(PCA_components)
```

```python

PCA_components.iloc[:, 0:1].to_csv('PCA_components.csv')
```

```python

ks = range(1, 10)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(PCA_components.iloc[:,:2])
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.title("Inertia VS # of Clusters")
plt.savefig("k-means_inertia.pdf")
plt.show()
```

```python
model = KMeans(n_clusters=5)
model.fit(PCA_components.iloc[:,:2])
labels = model.predict(PCA_components.iloc[:,:2])
labels
```

```python

plot_df = PCA_components.iloc[:,:2]
plot_df.columns = ['a', 'b']
```

```python
plot_df['label'] = labels
```

```python
groups = plot_df.groupby('label')

# Plot
fig, ax = plt.subplots()
ax.margins(0.05)
for name, group in groups:
    ax.plot(group.a, group.b, marker='o', linestyle='', ms=5, label=name)
ax.legend()
plt.title("Clustering Patients with K-means (K=2) ")
# plt.savefig("Kmeans_clusters.pdf")

plt.show()
```
